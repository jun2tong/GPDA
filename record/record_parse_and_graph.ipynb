```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
relevant_path = "."
included_extensions = ['test.txt']
file_names = [fn for fn in os.listdir(relevant_path)
              if any(fn.endswith(ext) for ext in included_extensions)]
param_file = open("./commands.txt")
data = []
for line in param_file.readlines():
    run = (str.split(line)[-1]).replace('\"', '')
    for fp in file_names:
        if run in fp:
            data.append([line, fp])
for datum in data:
    title = datum[0]
    seriesX = []
    seriesY = []
    for line in (open(datum[1])).readlines():
        seriesX.append(float(str.split(line)[1]))
        seriesY.append(float(str.split(line)[-1].replace("%",'')))
    print("Max accuracy:", max(seriesY))
    plt.plot(seriesX, seriesY)
    plt.title(datum[0])
    plt.show() 
```
